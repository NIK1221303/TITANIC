```python
#fare - median 
#age -ffill
#embarked -mode
#cabin - drop column
```

```python
df2= df.copy()
df2["Age"]=df2["Age"].fillna(method='ffill')
df2["Embarked"]=df2["Embarked"].fillna("S")
df2["Fare"]=df2["Fare"].fillna(14.454200)
df2=df2.drop('Cabin',axis=1)
df2.isnull().sum()
```

```python
df2.shape
```

```python
df2.head()
```

```python
#label encode sex,embarked

from sklearn.preprocessing import OrdinalEncoder

oe = OrdinalEncoder()
encoded = oe.fit_transform(df2[['Embarked','Sex']].values)
encoded_df = pd.DataFrame(encoded, columns=['Embarked','Sex'])
df3 = df2.drop(['Embarked','Sex'],axis=1)
df4=pd.concat([df3,encoded_df],axis=1)
df4 = df4.drop(columns=['Name','Ticket'])

df4.head()
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
train = df4[~df4["Survived"].isnull()]
test = df4[df4["Survived"].isnull()]
```

```python
train.isnull().sum()
```

```python
train.tail()
```

```python
test = test.drop(columns=['Survived'])
test.head()
```

```python
y_train = train["Survived"]
X_train = train.drop(columns=["Survived","PassengerId"])
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
model = DecisionTreeClassifier(random_state=42)
model.fit(X_train, y_train)
```

```python
from sklearn.model_selection import train_test_split

x_tr,x_va,y_tr,y_va = train_test_split(X_train,
                                       y_train,
                                       test_size=0.2,
                                       stratify=y_train,
                                       random_state=42)
```

```python
print("TEST:", x_tr.shape, y_tr.shape)
print("EVALUATION:", x_va.shape, y_va.shape)
```

```python
y_tr.value_counts()
```

```python
y_va.value_counts()

```

```python
model.fit(x_tr,y_tr)
```

```python
y_tr_pred = model.predict(x_tr)
y_va_pred = model.predict(x_va)
```

```python
from sklearn.metrics import accuracy_score
accuracy_tr = accuracy_score(y_tr,y_tr_pred)
accuracy_va = accuracy_score(y_va,y_va_pred)

print("TEST SCORE:",accuracy_tr)
print("EVALUATION SCORE:",accuracy_va)
```
